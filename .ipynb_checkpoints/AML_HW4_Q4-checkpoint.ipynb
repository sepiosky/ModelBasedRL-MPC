## **Q1**

دلیل تخمین تغییر در استیت به جای مدل استیت بعدی این است که ممکن است غالبا استیت فعلی و بعدی در رپرزنتییشن مورد استفاده بسیار نزدیک باشند و تفاوت تاثیر اعمال بر تاغییر این مقدار بسیار ناچیز باشد. در این صورت آموزش یک شبکه برای یادگیری این تفاوت بین اعمال در ایجاد استیت بعدی پیچیده خواهد شد. این مشکل هنگامی که تفاوت زمانی بیین اعمال کوچک است (گسسته سازی زمان پیوسته بسیار ریز دانه است) تشدید خواهد شد و یادگیری دقیق دایینامییک محیط و تفاوت تاثیر اعمال بر استیت ها سخت خواهد شد. به همی دللی یادگیری مستقییم تفاوت تاثیر اعمال در محیط به عنوان داینامیک موثر تر است. هم چنین از این مدل می توان به عنوان مدل معکوس محیط نیز استفاده کرد.

همانطور که در روش های آف پالیسی نیز مشکل دیستریبیوشن میسمچ وجود دارد، اینجا نیز اپدیت کردن بافر دیتا با دیتای ان پالیسی می تواند توزیع دیتاست را به توزیع استیت-اکشن ویزیتیشن پالیسی ای که از مدل استفاده می کند نزدیک کند و در نتیجه وزن دیتا پوینت ها در لاس پالیسی و داینامیک به واقعیت نزدیک بوده و دقت مدل ها بالاتر روند.

مشکل بهینه سازی اشاره شده یک این است که infinite-horizon دارد و ابجکتیو بهینه سازی احتمالا در مسیر جواب اپتیمال نامحدود خواهد شد. همچنین چون بهینه سازی این آبجکتیو به دلیل غیرخطی بودن داینامیک و ریوارد فانشکن دشوار است و به این منظور از تخمین های مناسب مسائل کنترل غیرخطی می توان استفاده کرد.
